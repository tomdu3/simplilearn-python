```python
from matplotlib import pylab
print(pylab.__version__)
```

```python
''' Use NumPy to generate random data
'''

import numpy as np

x = np.linspace(0, 10, 25)
y = x * x + 2
print(x)
print(y)
```

```python
'''It only takes 1 command to draw
'''
pylab.plot(x, y, 'r')  # 'r' stands for red
```

```python
''' Drawing a subgraph
'''

pylab.subplot(1, 2, 1)  # The contents of the brackets represent (rows, columns, indexes)
pylab.plot(x, y, 'r--')  # the third parameter here determines color and line style
pylab.subplot(1, 2, 2)
pylab.plot(y, x, 'g*-')
```

```python
# (1, 2, 1)  and (1,2,2) are significant if we want tiled subplots, without which both the graphs will be overlapped

pylab.subplot(1, 2, 1)  # The contents of the brackets represent (rows, columns, indexes)
pylab.plot(x, y, 'r--')  # the third parameter here determines color and line style
pylab.subplot(1, 2, 1)
pylab.plot(y, x, 'g*-')
```

# Operator Description

- `fig.add_axes()` - initializes subplot `a = fig.add_subplot(222)`
- `fig.b = plt.subplots(nrows=2, ncols=2)` - adds subplot
- `ax = plt.subplots(2, 3)` - creates subplot

```python
from matplotlib import pyplot as plt
```

```python
fig = plt.figure()

axis = fig.add_axes([0.5, 0.1, 0.8, 0.8])  # control the left, bottom, width, height of the canvas (from 0 to 1)

axis.plot(x, y, 'r')
```

```python
# again we can draw subgraphs

fig, axes = plt.subplots(nrows = 1, ncols = 2)  # submap is of 1 row, 2 columns
for ax in axes:
    ax.plot(x, y, 'r')
```

```python
# we can also draw a picture, or graph, inside another graph

fig = plt.figure()

# control the left, bottom, width, height of the canvas (from 0 to 1)
axes1 = fig.add_axes([0.1, 0.1, 0.8, 0.8])  # big axes
axes2 = fig.add_axes([0.2, 0.5, 0.4, 0.3])  # small canvas

axes1.plot(x,y, 'r')
axes2.plot(y,x, 'g')
```

```python
fig = plt.figure(figsize=(16, 9), dpi=300)  # New graphic object

fig.add_subplot()

plt.plot(x, y, 'r')
```

```python
fig, axes = plt.subplots()
# labels go before the title
axes.set_xlabel('x')
axes.set_ylabel('y')
axes.set_title('title')
plt.plot(x, y, 'r')
```

```python
ax.legend('label1', 'label2')
fig, axes = plt.subplots()
# labels go before the title
axes.set_xlabel('x-label')
axes.set_ylabel('y-label')
axes.set_title('title of graph')
axes.plot(x, x**2, 'r')
axes.plot(x, x**3, 'b')

axes.legend(['y = x**2', 'y = x**3'], loc=2)
```

```python
# in Matplotlib, you can set other properties such as line color, transparency, and more.

fig, axes = plt.subplots(dpi=150)
axes.plot(x, x**2, color='red', alpha=.3)
axes.plot(x, x+2, color='#1155dd', alpha=.3)
axes.plot(x, x+3, color='#15cc55', alpha=.3)
```

# Other 2D graphics

```python
n = np.array([0, 1, 2, 3, 4, 5])

fix, axes = plt.subplots(1,4, figsize=(16, 5))
axes[0].set_title("scatter")

axes[0].scatter(x, x + 0.25*np.random.randn(len(x)))
axes[1].set_title("step")

axes[1].step(n, n**2, lw=2)
axes[2].set_title("bar")

axes[2].bar(n, n**2, align="center", width=0.5, alpha=0.5)
axes[3].set_title("fill_between")

axes[3].fill_between(x, x**2, x**3, color="green", alpha=0.5)
```

```python
''' Draw a radar chart
'''

fig = plt.figure(figsize=(6, 6))
ax = fig.add_axes([0.0, 0.0, 0.6, 0.6], polar=True)
t = np.linspace(0, 2*np.pi, 100)
ax.plot(t, t*2, color='blue', lw=1)
N = 20
theta = np.arange(0.0, 2*np.pi, 2*np.pi/N)
radii = 10*np.random.rand(N)
width = np.pi/4*np.random.rand(N)
bars = ax.bar(theta, radii, width=width, bottom=0.0)

for r,bar in zip(radii, bars):
    bar.set_facecolor(plt.cm.jet(r/10.))
    bar.set_alpha(0.5)

plt.show()
```

```python
''' Draw a histogram
'''

n = np.random.randn(100000)
fig, axes = plt.subplots(1, 2, figsize=(12, 4))

axes[0].set_title("Default histogram")
axes[0].hist(n)

axes[1].set_title("Cumulative detailed histogram")
axes[1].hist(n, cumulative=True, bins=8)
```

```python
# Draw contour image

import matplotlib
import numpy as np
import matplotlib.cm as cm
import matplotlib.pyplot as plt

delta = 0.025
x = np.arange(-3.0, 3.0, delta)
y = np.arange(-2.0, 2.0, delta)
X, Y = np.meshgrid(x, y)
Z1 = np.exp(-X**2 - Y**2)
Z2 = np.exp(-(X - 1)**2 - (Y - 1)**2)
Z = (Z1 - Z2) * 2
```

```python
fig, ax = plt.subplots()
# im = ax.imshow(Z, interpolation='bilinear', cmap=cm.hot,
#                origin='lower', extent=[-3, 3, -2, 2],
#                vmax=abs(Z).max(), vmin=-abs(Z).max())

CS = ax.contour(X, Y, Z)

ax.clabel(CS, inline=1, fontsize=10)
ax.set_title('Contour Map')
plt.show()
```
