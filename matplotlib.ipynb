- it is used to draw 2D and 3D plots

- you can understand your data easily by visualizing it with the helpof matplotlib
- you can generate plots, histograms, bar charts and many other charts with just a few lines of code


Basic types of plots:
1. Bar Chart
2. Scatter Plot
3. Pie Chart
4. Histogram

```python
from matplotlib import pylab
print(pylab.__version__)
```

```python
''' Use NumPy to generate random data
'''

import numpy as np

x = np.linspace(0, 10, 25)
y = x * x + 2
print(x)
print(y)
```

```python
'''It only takes 1 command to draw
'''
pylab.plot(x, y, 'r')  # 'r' stands for red
```

```python
''' Drawing a subgraph
'''

pylab.subplot(1, 2, 1)  # The contents of the brackets represent (rows, columns, indexes)
pylab.plot(x, y, 'r--')  # the third parameter here determines color and line style
pylab.subplot(1, 2, 2)
pylab.plot(y, x, 'g*-')
```

```python
# (1, 2, 1)  and (1,2,2) are significant if we want tiled subplots, without which both the graphs will be overlapped

pylab.subplot(1, 2, 1)  # The contents of the brackets represent (rows, columns, indexes)
pylab.plot(x, y, 'r--')  # the third parameter here determines color and line style
pylab.subplot(1, 2, 1)
pylab.plot(y, x, 'g*-')
```

# Operator Description

- `fig.add_axes()` - initializes subplot `a = fig.add_subplot(222)`
- `fig.b = plt.subplots(nrows=2, ncols=2)` - adds subplot
- `ax = plt.subplots(2, 3)` - creates subplot

```python
from matplotlib import pyplot as plt
```

```python
fig = plt.figure()

axis = fig.add_axes([0.5, 0.1, 0.8, 0.8])  # control the left, bottom, width, height of the canvas (from 0 to 1)

axis.plot(x, y, 'r')
```

```python
# again we can draw subgraphs

fig, axes = plt.subplots(nrows = 1, ncols = 2)  # submap is of 1 row, 2 columns
for ax in axes:
    ax.plot(x, y, 'r')
```

```python
# we can also draw a picture, or graph, inside another graph

fig = plt.figure()

# control the left, bottom, width, height of the canvas (from 0 to 1)
axes1 = fig.add_axes([0.1, 0.1, 0.8, 0.8])  # big axes
axes2 = fig.add_axes([0.2, 0.5, 0.4, 0.3])  # small canvas

axes1.plot(x,y, 'r')
axes2.plot(y,x, 'g')
```

```python
fig = plt.figure(figsize=(16, 9), dpi=300)  # New graphic object

fig.add_subplot()

plt.plot(x, y, 'r')
```

```python
fig, axes = plt.subplots()
# labels go before the title
axes.set_xlabel('x')
axes.set_ylabel('y')
axes.set_title('title')
plt.plot(x, y, 'r')
```

```python
ax.legend('label1', 'label2')
fig, axes = plt.subplots()
# labels go before the title
axes.set_xlabel('x-label')
axes.set_ylabel('y-label')
axes.set_title('title of graph')
axes.plot(x, x**2, 'r')
axes.plot(x, x**3, 'b')

axes.legend(['y = x**2', 'y = x**3'], loc=2)
```

```python
# in Matplotlib, you can set other properties such as line color, transparency, and more.

fig, axes = plt.subplots(dpi=150)
axes.plot(x, x**2, color='red', alpha=.3)
axes.plot(x, x+2, color='#1155dd', alpha=.3)
axes.plot(x, x+3, color='#15cc55', alpha=.3)
```

```python
fix, ax = plt.subplots(dpi=100)

# Line width

ax.plot(x, x+1, color='blue', linewidth=.25)
ax.plot(x, x+2, color='blue', linewidth=1)
ax.plot(x, x+3, color='blue', linewidth=1.5)
ax.plot(x, x+4, color='blue', linewidth=2)
```

```python
fix, ax = plt.subplots(dpi=100)

# Line width

line, = ax.plot(x, x+1, color='blue', lw=.25, linestyle='-')
line.set_dashes([5, 10, 15, 10])

line, = ax.plot(x, x+2, color='blue', lw=1, linestyle='-.')
line.set_dashes([5, 10, 15, 3])

line, = ax.plot(x, x+3, color='blue', lw=1.5, linestyle=':')
line.set_dashes([5, 10, 15, 10])

# ax.plot(x, x+4, color='blue', lw=2)

line, = ax.plot(x, x+4, color='black', lw=1.5)
line.set_dashes([5, 10, 30, 30])
```

```python
fix, ax = plt.subplots(dpi=100)
ax.plot(x, x+1, color='blue', marker='o')
ax.plot(x, x+2, color='blue', marker='+')
ax.plot(x, x+3, color='blue', marker='s')
ax.plot(x, x+4, color='blue', marker='1')
ax.plot(x, x+5, color='blue', marker='o', markersize=2)
ax.plot(x, x+6, color='blue', marker='o', markerfacecolor='red')
ax.plot(x, x+7, color='blue', marker='s', markersize=15, markerfacecolor='red')
 
```

```python
''' Set the canvas grid and axis range
'''

fig, Axes = plt.subplots(1, 2, figsize =(10, 5))

Axes[0].plot(x, x**2, x, x**3, lw=2)
Axes[0].grid(True)

Axes[1].plot(x, x**2, x, x**3, lw=2)
Axes[1].set_ylim([0, 60])
Axes[1].set_xlim([2, 5])
Axes[1].grid(True)
```
